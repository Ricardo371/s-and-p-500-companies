```
#Then I run a for loop for all the tickers in the constituents list, to get the CSV from each one.
#After getting the data, I left only the closing price, and converted it to CSV with the ticker as name.

for ticker in constituents['Symbol']:
    stock = yf.download(ticker, start="2010-01-01", end="2021-02-15")
    stock = stock['Close']
    stock.to_csv(ticker+'.csv')
```

```
#Here I run a for loop for each CSV, I append them to the list li and then concatenate it to the df prices
li = []
for ticker in constituents['Symbol']:
  df= pd.read_csv(ticker+'.csv', index_col=0,header=0)
  li.append(df)
  
  prices = pd.concat(li, axis=1, ignore_index=True)
```

```
#Here I make a list of the constituents for later use.
Names = [constituents['Symbol']]
```

```
#Here I label the columns of our prices df with the list names previously made.
prices.columns = Names
prices.drop(prices.tail(2).index,inplace=True)
```

```
#Then finally I convert the df to a csv file for storage and later use.
prices.to_csv('prices_S&P500.csv')
```
